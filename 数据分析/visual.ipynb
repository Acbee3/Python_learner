```python
import pandas as pd

from matplotlib import pyplot as plt
```

```python
# 创建李健的数据
li_jian = pd.Series([80, 85, 89, 91, 88, 95],
                    index=['2月', '3月', '4月', '5月', '6月', '7月'])
# 查看li_jian
li_jian
```

```python
# 设置中文字体
plt.rcParams['font.family'] = ['SimSun']
# 绘制单条折线图
li_jian.plot(kind='line', figsize=(7, 7), title='李健月考成绩')
```

```python
# 绘制李健成绩的折线图
li_jian.plot(kind='line', figsize=(10, 10), xticks=[0], yticks=[100], rot=30)
```

```python
# 导入数据
my_data = pd.read_csv('./课程素材/工作/test_data.csv', encoding='utf-8')
# 查看my_data
my_data
```

```python
# 获取行业的频率分布
profession = my_data['品牌'].value_counts()/my_data['品牌'].value_counts().sum()
# 绘制行业频率分布的饼图
profession.plot(kind='pie', autopct='%.2f%%', figsize=(7, 7), title='行业频率分布图', label='')
```

```python
mask_data_clean = pd.read_csv('./课程素材/工作/mask_data_clean.csv', encoding='utf-8')
mask_data_clean
```

```python
# 获取每月销售额总数
sales_income = mask_data_clean.groupby('月份')['销售额'].sum()
sales_income
```

```python
# 获取各月总订单量
order_number = mask_data_clean.groupby(['月份'])['订单量'].sum()
order_number
```

```python
# 获取每月平均单价，存储在变量 month_price 中
month_price = mask_data_clean.groupby(['月份'])['单价'].mean()
# 查看数据
month_price
```

```python
# 获取各月各省总订单量
month_order1 = mask_data_clean.groupby(['月份', '省'])['订单量'].sum()
# 查看数据
month_order1
```

```python
# 画出各月总销售额的折线图
sales_income.plot(kind='line', figsize=(7, 7), title='各月总销售额趋势图')
```

```python
# 画出各月总订单量的折线图
order_number.plot(kind='line', title='各月总订单量')
```

```python
# 画出各月平均单价的折线图
month_price.plot(kind='line', title='各月平均单价')
```

```python
# 将month_order1转换为DataFrame，存储在month_order1_df中
month_order1_df = month_order1.unstack()
# 查看month_order1_df中的数据
month_order1_df
```

```python
month_order1_df.plot(kind='line', title='各月各省总订单量趋势图')
```
